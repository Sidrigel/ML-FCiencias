```python
from numpy import*
from pylab import*
from scipy import*
```

```python
x=linspace(-5,5,1000)
```

```python
def normal(x,m,s):
    return (1/sqrt(2*pi*s**2)*exp(-(1/2*s**2)*(x-m)**2))
```

```python
y=normal(x,0,1)
```

```python
def ascensoColina(lista,indice):
    if lista[indice]>lista[indice+1]:
        return lista[indice],indice
    else:
        while(lista[indice]<=lista[indice+1]):
            indice=indice+1
        return lista[indice],indice    
```

```python
f=ascensoColina(y,5)
punto0=x[f[1]]
punto1=f[0]
```

```python
plot(x,y)
plot(punto0,punto1,marker='.',color='red')
```

```python
def fleo(x):
    return((-x-5)*(x-1)*(x+3)*(x-4))
```

```python
xleo=linspace(-6,4.1,100)
yleo=fleo(xleo)
```

```python
f=ascensoColina(yleo,5)
punto0=xleo[f[1]]
punto1=f[0]
```

```python
plot(xleo,yleo)
plot(punto0,punto1,marker='.',color='red')
```
